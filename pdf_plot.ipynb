```python
import numpy as np
import matplotlib.pyplot as plt

means = [0.4, 0.5, 0.7]
sigma = 1
m = len(means)  # 3

# Compute k
k = 1 / (m * sigma * np.sqrt(2 * np.pi))

# Range
x_val = np.arange(0, 1.1, 0.01)
p_val = []

# Calculating p(x) values
for x in x_val:
    sum = 0
    for mean in means:
        sum += np.exp(-((x - mean) ** 2) / (2 * sigma ** 2))
    p = k * sum
    p_val.append(p)

# Plot
plt.plot(x_val, p_val, label='pdf')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Probability Density Function')
plt.legend()
plt.show()
```
